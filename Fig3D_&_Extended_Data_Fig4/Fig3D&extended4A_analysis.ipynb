```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# import excel
filename = 'Fig3D&extended4A_data.xlsx'
xl = pd.ExcelFile(filename)
print(xl.sheet_names)
```

```python
# import sp (-oriT) data
s1_df = xl.parse('sp (-oriT)', index_col = 0, header = 0)
s1_df = s1_df.reset_index()
s1_df = pd.melt(s1_df, id_vars='day', value_name = 'Fp')
s1_df.insert(1, 'sample', np.array(['sp (-oriTF)']*24))
s1_df.rename(columns = {'variable':'Replicate',}, inplace = True)
s1_df
```

```python
# do the same for the rest 5 samples

s2_df = xl.parse('sp (+oriT)', index_col = 0, header = 0).reset_index()
s2_df = pd.melt(s2_df, id_vars='day', value_name = 'Fp')
s2_df.insert(1, 'sample', np.array(['sp (+oriTF)']*24))
s2_df.rename(columns = {'variable':'Replicate',}, inplace = True)

s3_df = xl.parse('NT (+oriT)', index_col = 0, header = 0).reset_index()
s3_df = pd.melt(s3_df, id_vars='day', value_name = 'Fp')
s3_df.insert(1, 'sample', np.array(['NT (+oriTF)']*24))
s3_df.rename(columns = {'variable':'Replicate',}, inplace = True)

s4_df = xl.parse('Msp1', index_col = 0, header = 0).reset_index()
s4_df = pd.melt(s4_df, id_vars='day', value_name = 'Fp')
s4_df.insert(1, 'sample', np.array(['Msp1']*24))
s4_df.rename(columns = {'variable':'Replicate',}, inplace = True)

s5_df = xl.parse('Msp2', index_col = 0, header = 0).reset_index()
s5_df = pd.melt(s5_df, id_vars='day', value_name = 'Fp')
s5_df.insert(1, 'sample', np.array(['Msp2']*24))
s5_df.rename(columns = {'variable':'Replicate',}, inplace = True)

s6_df = xl.parse('Msp3', index_col = 0, header = 0).reset_index()
s6_df = pd.melt(s6_df, id_vars='day', value_name = 'Fp')
s6_df.insert(1, 'sample', np.array(['Msp3']*24))
s6_df.rename(columns = {'variable':'Replicate',}, inplace = True)
```

```python
# import calculated mean data
mean_df = xl.parse('mean', index_col = 0, header = 0).reset_index()
mean_df = pd.melt(mean_df, id_vars='day', value_name = 'Fp')
sample_ids = np.repeat(np.array(['sp (-oriTF)', 'sp (+oriTF)', 'NT (+oriTF)', 'Msp1', 'Msp2', 'Msp3']),6)
mean_df.insert(1, 'sample', sample_ids)
mean_df
```

```python
individual_df = pd.concat([s1_df, s2_df, s3_df, s4_df, s5_df, s6_df])
individual_df = individual_df.reset_index(drop=True)
individual_df
```

```python
fig, axes = plt.subplots(1,1, figsize=(9, 2), sharey = True, sharex = True, dpi=200)
plt.rcParams['font.sans-serif'] = "Myriad Pro"
# fig.tight_layout()

sns.lineplot(data = mean_df[(mean_df['sample']!='Msp2')&(mean_df['sample']!='Msp3')], 
            x = 'day', y='Fp', hue = 'sample', ax = axes)

sns.scatterplot(data = individual_df[(individual_df['sample']!='Msp2')&(individual_df['sample']!='Msp3')], x = 'day', y = 'Fp', hue = 'sample', alpha = 0.5, ax = axes)

# sns.move_legend(axes, "upper left", bbox_to_anchor=(1, 1))
axes.set_yscale('log')
axes.set_ylim([1e-2, 2])
axes.tick_params(axis='both', which='major', labelsize=15)


handles, labels = axes.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
# axes.legend(by_label.values(), by_label.keys(),bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., prop = { "size": 15 },)
axes.legend_=None
axes.set_xlabel('Day', fontsize = 15 )
axes.set_ylabel('Plasmid abundance (Fp)', fontsize = 12)
plt.savefig('fig3d.svg', format='svg',bbox_inches='tight', pad_inches=0.1)

plt.show()
```

```python
fig, axes = plt.subplots(1,1, figsize = (10,4), sharey = True, sharex = True, dpi=200)
plt.rcParams['font.sans-serif'] = "Myriad Pro"


sns.lineplot(data = mean_df, 
            x = 'day', y='Fp', hue = 'sample', ax = axes)

sns.scatterplot(data = individual_df, x = 'day', y = 'Fp', hue = 'sample', alpha = 0.5, ax = axes)

sns.move_legend(axes, "upper left", bbox_to_anchor=(1, 1))
axes.set_yscale('log')
axes.set_ylim([1e-2, 2])
axes.tick_params(axis='both', which='major', labelsize=15)


handles, labels = axes.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
axes.legend(by_label.values(), by_label.keys(),bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., prop = { "size": 15 },)

axes.set_xlabel('Day', fontsize = 20 )
axes.set_ylabel('Plasmid abundance (Fp)', fontsize = 20)
plt.savefig('extended_fig4a.svg', format='svg',bbox_inches='tight', pad_inches=0.1)

plt.show()
```
