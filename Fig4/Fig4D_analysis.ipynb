```python
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
filename = 'Fig4D_data.xlsx'
xl = pd.ExcelFile(filename)
df = xl.parse('Fig4D', index_col = 0, header = 0).reset_index()
df
```

```python
fig, ax = plt.subplots(1,1, figsize = (5,1.8), dpi = 200)

sns.lineplot(data = df, marker = 'o', markersize = 7,
             x = 'Hours', y = 'Green (a.u.)', ax = ax, color = '#14A201', lw = 2)
right_ax = ax.twinx()
sns.lineplot(data = df, marker = 'o', markersize = 7, 
                 x = 'Hours', y = 'Red (a.u.)', ax = right_ax, color = '#CA02CA', lw = 2)

# <0 kan
ax.axvspan(xmin=-5, xmax=0, facecolor='#FFFF85', alpha=0.3) # kan  
ax.axvspan(xmin=0, xmax=11, facecolor='#FEC2FE', alpha=0.3) # rha  
ax.axvspan(xmin=11, xmax=32.5, facecolor='#b7efc5', alpha=0.3) # ara  
ax.axvspan(xmin=32.5, xmax=48, facecolor='#FFFF85', alpha=0.3) # kan  


ax.set_ylabel('GFP (a.u.)', color = '#14A201', fontsize = 12)
right_ax.set_ylabel('RFP (a.u.)', color = '#CA02CA', fontsize = 12)
ax.set_xlim([-5, 50])

plt.savefig('fig4d.svg', format='svg',bbox_inches='tight', pad_inches=0.1)


plt.show()
```
